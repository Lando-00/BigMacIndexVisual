```python
import pandas as pd
import numpy as np
```

```python
BigMac = pd.read_csv('big_macMap.csv')
BigMac.head(5)
```

```python
BigMac.dtypes
```

```python
# change date attribute to time format
BigMac['date'] = pd.to_datetime(BigMac['date'])
BigMac.dtypes
```

```python
# Split dataframe into different Years
BigMac_YearsSplit = []

for year in range(2000, 2021):
    BigMac_YearsSplit.append(BigMac[BigMac['date'].dt.year == year])
```

```python
BigMac_YearsSplit[0].head(5)
```

```python
len(BigMac_YearsSplit)
```

```python
# Write each time df into csv file with a naming scheme
i = 2000
for df in BigMac_YearsSplit:
    df.index('name')
    df.to_csv('big_mac{}'.format(i))
    i+=1
```
